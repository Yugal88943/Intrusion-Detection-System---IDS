```python
y = df['class']
```

```python
X['protocol_type'].unique() #Internet Control Message Protocol-icmp
```

```python
from sklearn.preprocessing import OneHotEncoder

X_encoded = pd.get_dummies(
    X,
    columns=['protocol_type', 'service', 'flag']
)
```

```python
columns = X_encoded.columns
```

```python
X_encoded.shape
```

```python
y.unique()
```

```python
y = y.apply(lambda attack: 0 if attack == "normal" else 1)
```

```python
y.shape
```

```python
y.head()
```

```python
y.unique()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X_encoded)
```

```python
type(X_scaled)
```

Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size= 0.2,
    random_state= 42
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model = LogisticRegression(max_iter=1000, class_weight='balanced')
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
probabilities = model.predict_proba(X_test)
```

```python
probabilities[:5]
```

```python
attack_probabilities = probabilities[:, 1]
```

```python
attack_probabilities
```

```python
custom_prediction = (attack_probabilities > 0.4).astype(int)
```

```python
custom_prediction
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, custom_prediction)
print("Accuracy: ", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, custom_prediction)
print("Confusion Matrix: \n", cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Greens'
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Logistic Regression Confusion Matix")
plt.show()
```

```python
from sklearn.metrics import recall_score
```

```python
recall = recall_score(y_test, custom_prediction)
print("Recall: ", recall)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
fpr, tpr, thresholds = roc_curve(y_test, attack_probabilities)
```

```python
lr_auc = auc(fpr, tpr)
print("Logistic Regression AUC: ", lr_auc)
```

```python
import matplotlib.pyplot as plt

plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

Artificial Neural Network

```python
from sklearn.neural_network import MLPClassifier
```

```python
ann = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
```

```python
ann.fit(X_train, y_train)
```

```python
ann_prediction = ann.predict(X_test)
```

```python
ann_accuracy = accuracy_score(y_test, ann_prediction)
print("ANN Accuracy: ", ann_accuracy)
```

```python
ann_cm = confusion_matrix(y_test, ann_prediction)
print("ANN Confusion Matrix: \n", ann_cm)
```

```python
ann_recall = recall_score(y_test, ann_prediction)
print("ANN Recall: ", ann_recall)
```

```python
import seaborn as sns

plt.figure(figsize=(6,5))

sns.heatmap(
    ann_cm,
    annot=True,
    fmt='d',
    cmap='Blues'
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("ANN Confusion Matix")
plt.show()
```

```python
models = [
    "LR",
    "ANN"
]

recall_values = [
    97.53,
    99.58
]

plt.figure(figsize=(8,5))
plt.bar(models, recall_values)
plt.ylabel("Recall (%)")
plt.title("Recall Comparison of Models")
plt.ylim(90,100)
plt.show()
```

```python
models = [
    "LR",
    "ANN"
]

false_negatives = [
    291,
    49
]

plt.figure(figsize=(8,5))
plt.bar(models, false_negatives)
plt.ylabel("False Negatives")
plt.title("Missed Attack Comparison")
plt.show()
```

```python
ann_probabilities = ann.predict_proba(X_test)[:,1]
```

```python
ann_fpr, ann_tpr, _ = roc_curve(y_test, ann_probabilities)
```

```python
ann_auc = auc(ann_fpr, ann_tpr)
print("ANN AUC: ", ann_auc)
```

```python
plt.figure(figsize=(6,5))

plt.plot(ann_fpr, ann_tpr)
plt.title("ANN ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positve Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt

metrics = [
    "Accuracy",
    "Recall",
    "False Negatives",
    "False Positives",
    "AUC"
]

lr_values = [
    "97.43%",
    "97.53%",
    "291",
    "355",
    "0.9959"
]

ann_values = [
    "99.67%",
    "99.58%",
    "49",
    "34",
    "0.9994"
]

table_data = []

for i in range(len(metrics)):
    table_data.append([
        metrics[i],
        lr_values[i],
        ann_values[i]
    ])

fig, ax = plt.subplots(figsize=(12, 4))

ax.axis('off')

table = ax.table(
    cellText=table_data,
    colLabels=[
        "Metric",
        "Logistic Regression",
        "ANN"
    ],
    cellLoc='center',
    colLoc='center',
    loc='center'
)

table.auto_set_font_size(False)
table.set_fontsize(11)

table.scale(1.3, 2.2)

for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold', fontsize=12)
        cell.set_height(0.15)

    cell.PAD = 0.08

plt.title(
    "Performance Comparison of Logistic Regression and ANN",
    fontsize=15,
    pad=20
)

plt.show()
```
